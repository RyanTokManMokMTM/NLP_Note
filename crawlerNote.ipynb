```python
url = "https://movies.yahoo.com.tw/movie_intheaters.html"
r = requests.get(url)
soup = BeautifulSoup(r.text)
```

```python
#get first one 
print(soup.find("div",class_="release_movie_name").find("a",class_="gabtn").text.strip())
```

```python
#get all movie name
data = soup.findAll("div",class_="release_movie_name")
for d in data:
  print(d.find("a",class_="gabtn").text.strip())
```

```python
url = "https://movies.yahoo.com.tw/movie_thisweek.html"
r = requests.get(url)
soup = BeautifulSoup(r.text)

name = soup.findAll("div",class_="release_movie_name")
image = []
infos = []
for d in soup.findAll("div",class_="release_info_text"):
  dx = {}
  dx["ChineseName"] = d.find("div",class_="release_movie_name").a.text.strip()
  dx["EnglishName"] = d.find("div",class_="release_movie_name").find("div",class_="en").a.text.strip()
  dx["RealseDate"] = d.find("div",class_="release_movie_time").text.split("上映日期 ： ")[1]
  dx["Expect"] = 0
  if d.find("div",class_="levelbox"):
      dx['Expect'] = d.find("div",class_="levelbox").span.text
  infos.append(dx)


for d in soup.findAll("div",class_="release_foto"):
  img = {}
  img["ImgURL"] = d.find("img")['src']
  image.append(img)

res = []
for i,j in list(zip(image,infos)):
  res.append(dict(list(i.items())+list(j.items())))

res
```

# DownLoad data and Img

```python
import pandas as pd
df = pd.DataFrame(res)
df.to_csv('movie.csv',encoding="utf_8_sig")

df
```
